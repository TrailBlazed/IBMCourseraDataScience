## Finding the time of the day when most accident occur

```python
l={range(4,8):'early morning',range(8,12): 'morning',range(12,16):'afternoon',
   range(16,20):'evening', range(16,0):'midnight',range(0,4):'late night'} 
```

```python
{ l[key] for key in l if 6 in key}
```

```python
df['Time']=df['Time'].replace('',np.nan)
```

```python
from datetime import datetime
df['Time24']=pd.to_datetime(df['Time']).dt.strftime('%H:%M:%S')
```

```python
df.to_csv(r'C:\Users\tskir\Documents\certification\IBM\project\Data-Collisions1.csv')
```

```python
df['Hours']=pd.to_datetime(df['Time24'], format='%H:%M:%S').dt.hour
```

```python
df.to_csv(r'C:\Users\tskir\Documents\certification\IBM\project\Data-Collisions1.csv')
```

```python
l={range(4,8):'early morning',range(8,12): 'morning',range(12,16):'afternoon',
   range(16,20):'evening', range(16,0):'midnight',range(0,4):'late night'} 


df['TimeRange'] = df['Hours'].apply(lambda x: next((v for k, v in l.items() if x in k), 0))

```

```python
df['TimeRange']=df['TimeRange'].replace(0,"Unknown")
df['Hours']=df['Hours'].replace(np.NaN,"Unknown")
```

```python
df.to_csv(r'C:\Users\tskir\Documents\certification\IBM\project\Data-Collisions1.csv')
```

## The time range of a day when most accidents occur

```python

y = df['TimeRange'].value_counts(ascending=False)
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(y.index, y, height=0.75, color="Green")
plt.title('Count of Accidents')
plt.xlabel('Count')
plt.ylabel('Time')
_, xmax = plt.xlim()
plt.xlim(0, xmax+300)
for i, v in enumerate(y):
    ax.text(v + 100, i, str(v), color='black', fontweight='bold', fontsize=14, ha='left', va='center')
plt.show()
```

## Heatmap on a map

```python
from folium import plugins
from folium.plugins import HeatMap


map_hooray = folium.Map(location=[47.69034175,-122.3290808],
                    zoom_start = 13) 

# Ensure you're handing it floats
df['X'] = df['X'].astype(float)
df['Y'] = df['Y'].astype(float)

# Filter the DF for rows, then columns, then remove NaNs
heat_df = df[df['TimeRange']=='afternoon'] # Reducing data size so it runs faster
heat_df = heat_df[['X', 'Y']]
heat_df = heat_df.dropna(axis=0, subset=['X','Y'])

# List comprehension to make out list of lists
heat_data = [[row['X'],row['Y']] for index, row in heat_df.iterrows()]

# Plot it on the map
hm_wide = HeatMap( list(zip(heat_df.Y.values, heat_df.X.values)),
                     min_opacity=0.2,
                     radius=13, blur=15,
                     max_zoom=1
                 )
map_hooray.add_child(hm_wide)
# Display the map
map_hooray
```
